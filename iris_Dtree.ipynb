```python
import numpy as np
import pandas as pd
import sklearn.datasets as datasets

iris=datasets.load_iris()
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head(5)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
y=iris.target
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df,y, train_size=0.7, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(x_train, y_train)
```

```python
dtree.score(x_test, y_test)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import tree
fig=plt.figure(figsize=(45,40))
x=tree.plot_tree(dtree,feature_names=iris.feature_names,class_names=iris.target_names,filled=True,rounded=True,fontsize=30)
```

```python
y_pred=dtree.predict(x_test)
```

```python
y_pred
```

```python
result=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
```
